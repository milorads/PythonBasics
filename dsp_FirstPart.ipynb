```python
cos_sig = thinkdsp.CosSignal(freq=440, amp=1.0,offset=0)
sin_sig = thinkdsp.SinSignal(freq=880, amp=0.5, offset=0)
mix = sin_sig+cos_sig
```

```python
wave = mix.make_wave(duration=0.5, start=0, framerate=11025)
from IPython.display import Audio
audio = Audio(data=wave.ys, rate=wave.framerate)
```

```python
wave.make_audio()
```

```python
audio
```

```python
period = mix.period
segment = wave.segment(start=0, duration=period+0.003)
segment.plot()
```

```python
filename = 'temp.wav'
wave.write(filename)
thinkdsp.play_wave(filename=filename, player='aplay')
```
