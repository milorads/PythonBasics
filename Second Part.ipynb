# Discrete Fourier Transformation and Fast Fourier Transformation of the sine wave

```python
n = sp.arange(16)
x = sin(2*sp.pi*n*0.125)
# adding a DC component
# x = sin(2*sp.pi*n*0.125)+1
plotter.stem(n,x)
```

```python
X = sp.fft(x)
```

```python
plotter.clf
w = sp.linspace(0, sp.pi*2, len(x)+1)
w = w[:-1]
plotter.stem(w, sp.absolute(X))
```

```python
import numpy as np
n = sp.arange(32)
xx = sp.cos(2*sp.pi*n*0.125)
XX = sp.fft(xx)
w = sp.linspace(0, sp.pi*2,32)
plotter.stem(np.real(w), np.real(XX))
#extracted the Real part to avoid the warning that tells us that casting values to real discards the imaginary part
#(since thats what we want)
```
