```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import specgram
%matplotlib inline
```

```python
import scipy as sp
```

```python
t = sp.linspace(-10,10,2001)
```

```python
x = sp.cos(2*sp.pi*t*5)
x[len(x)/2:] = sp.cos(sp.pi*2*t[len(x)/2:]*12)
x += sp.cos(2*sp.pi*t*20)
```

```python
plt.plot(t,x)
```

```python
specgram(x, NFFT = 256, Fs = 100)
```
