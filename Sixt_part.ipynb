```python
#3. zadatak

Fs = 50
#minimalna frekv. odabiranja je dvostruko veca od maksimalne frekv.signala, ali kako je signal pri Fs=50 prilicno "izlomljen"
#koristicemo Fs = 100

t=sp.arange(0,3,1/Fs)

f = sp.cos(2*sp.pi*5*t) + sp.cos(2*sp.pi*10*t) + sp.cos(2*sp.pi*19*t)

plt.plot(t,f)
```

```python
#3. zadatak 1.1

b, a = signal.butter(4, [0.38, 0.42], btype = 'bandpass')

w, j = signal.freqz(b, a)

F = signal.lfilter(b, a, f)

plt.subplot(311)
plt.plot(w * (Fs/2) / sp.pi, sp.absolute(j))
plt.subplot(312)
plt.plot(f)
plt.subplot(313)
plt.plot(F)
```

```python
X=sp.fft(f)

Ft = sp.linspace(0,Fs, len(X)+1)
Ft = Ft[:-1]

Y=sp.fft(F)

Fft = sp.linspace(0,Fs, len(Y)+1)
Fft = Fft[:-1]

plt.subplot(211)
plt.plot(Ft, sp.absolute(X))
plt.subplot(212)
plt.plot(Fft, sp.absolute(Y))
```
