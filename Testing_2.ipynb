```python
#Zadatak 1
import thinkdsp
import thinkplot
import numpy as np
%matplotlib inline
```

```python
sonata = thinkdsp.read_wave('sonata.wav')
sonata.make_audio()
```

```python
sonata.make_audio()
sonata.normalize()
sonata.apodize()
sonata.plot()
```

```python
start = 3.0
duration = 0.5
segment = sonata.segment(start,duration)
segment.plot()
```

```python
spectrum = segment.make_spectrum()
spectrum.plot(high=4500)
```

```python
#spectrum.low_pass(500)
#spectrum.high_pass(500)
spectrum.band_stop(100,800)
spectrum.plot(high=4500)
```

```python
filtered = spectrum.make_wave()
filtered.make_audio()
```

```python
#Zadatak 2
u1Signal = thinkdsp.CosSignal(freq=50, amp=1.0, offset=0)
u2Signal = thinkdsp.SinSignal(freq=100, amp=0.8, offset=0)
u3Signal = thinkdsp.CosSignal(freq=150, amp=0.6, offset=0)
u4Signal = thinkdsp.SinSignal(freq=200, amp=0.4, offset=0)
centSignal = thinkdsp.CosSignal(freq=250, amp=0.2, offset=0)
d4Signal = thinkdsp.SinSignal(freq=300, amp=0.4, offset=0)
d3Signal = thinkdsp.CosSignal(freq=350, amp=0.6, offset=0)
d2Signal = thinkdsp.SinSignal(freq=400, amp=0.8, offset=0)
d1Signal = thinkdsp.CosSignal(freq=450, amp=1.0, offset=0)
```

```python
signal  = u1Signal + u2Signal + u3Signal + u4Signal + centSignal + d4Signal + d3Signal + d2Signal + d1Signal
signal.plot()
```

```python
wave = signal.make_wave()
spectrum = wave.make_spectrum()
spectrum.plot(high = 500)
```
